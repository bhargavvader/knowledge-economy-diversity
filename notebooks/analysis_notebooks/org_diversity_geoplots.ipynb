### Org Diversity Analysis

We will first look at org diversity, for some big orgs, as well as the lowest and highest.

```python
import json
```

```python
import numpy as np
import pandas as pd
```

```python
with open('../www_data/org_diversity.json') as json_file:
    org_diversity = json.load(json_file)
```

```python
org_avg_diversity = {}
```

```python
for org in org_diversity:
    try:
        org_avg_diversity[org] = org_diversity[org][0][0]
    except TypeError:
        if org_diversity[org] != 0:
            org_avg_diversity[org] = org_diversity[org]
```

```python
def find_entities(vectors, min_papers=100, kind=""):
    avg = []
    for vec in vectors:
        if kind in vec.lower():
            print(vec, vectors[vec])
            avg.append(vectors[vec])
    print("Average " + str(np.mean(avg)))
```

```python
find_entities(org_avg_diversity, kind="amazon")
```

```python
find_entities(org_avg_diversity, kind="microsoft")
```

```python
find_entities(org_avg_diversity, kind="intuit")
```

```python
msa = geopandas.read_file('../www_data/cb_2018_us_csa_500k/cb_2018_us_csa_500k.shp')
```

```python
columns = ["MSA", "Pop", "Count R", "Count T", "Count J", "Uni Count","Diff J", "Diff R", "Diff T"]
```

```python
msa_info = pd.read_csv("../www_data/msa_info_plot.csv", names=columns)
```

```python
msa_info 
```

```python
msa_info = msa_info[msa_info.Pop != 0]
```

```python
msa_info
```

```python
msa
```

```python
from thefuzz import process
```

```python
from thefuzz import fuzz
```

```python
msa_map = {}
```

```python
for index, row in msa.iterrows():
    name = row["NAME"]
    options = []
    for index_, row_ in msa_info.iterrows():
        try:
            name_, _ = row_["MSA"].split("Metro")
        except:
            name_, _ = row_["MSA"].split("Micro")
        options.append(name_)
    match, number = process.extractOne(name, options, scorer=fuzz.token_sort_ratio)
    if number > 70:
        msa_map[name] = match
```
